ATIVIDADE 1

```python
nome = " MARIA EDUARDA ALMEIDA "
email = " MARIA.ALMEIDA@EMPRESA.COM "

#1 -remover espaços extras
nome = nome.strip()
email = email.strip()
print("nome sem espaços:", nome)
print('email sem espaços:', email)

#2 - converter para minusculo
nome1 = nome.lower()
email2 = email.lower()
print('nome minusculo:', nome1, email2)

#3 - substituir o email

email = email.replace('empresa.com', 'gmail.com')
print('email tratado:', email)

#4 - quantidade de caracteres
quantidade = len(nome)
print('quantidade de caracteres:', quantidade)

#5 - ultimo/primeiro
ultima_letra = nome[-1]
primeira_letra = nome[0]
print('ultima letra:', ultima_letra)
print('primeira letra:', primeira_letra)

```
